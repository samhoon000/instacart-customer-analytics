```python
from sqlalchemy import create_engine

engine = create_engine(
    "mysql+pymysql://root:@localhost/instacart"
)
```

```python
aisles.to_sql(
    "aisles",
    engine,
    if_exists="replace",
    index=False
)
```

```python
departments.to_sql(
    "departments",
    engine,
    if_exists="replace",
    index=False
)

products.to_sql(
    "products",
    engine,
    if_exists="replace",
    index=False
)

orders.to_sql(
    "orders",
    engine,
    if_exists="replace",
    index=False,
    chunksize=50000
)
```

```python
for start in range(0, len(order_products_prior), 5000):
    chunk = order_products_prior.iloc[start:start+5000]

    chunk.to_sql(
        "order_products_prior",
        engine,
        if_exists="append" if start > 0 else "replace",
        index=False,
        method="multi"
    )

    print(f"Inserted {start + len(chunk):,} rows")
```
